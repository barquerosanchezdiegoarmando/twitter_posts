```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
com_ven = pd.read_excel("Compra&Venta.xlsx")
```

```python
data = com_ven.copy()
data.head()
```

```python
index = pd.date_range('1/1/2021', periods=957, freq='d')
data.set_index(index, inplace = True )
```

```python
data.drop("Fecha", axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.shape
```

```python
data.tail()
```

```python
plt.figure(figsize = (20,7))
plt.plot(data.index, data['TIPO CAMBIO COMPRA'])
plt.plot(data.index, data['TIPO DE CAMBIO VENTA'])
plt.xlabel('Periodicidad Diaria')
plt.ylabel('Tipo de cambio')
plt.title('Serie de tiempo')
plt.show()
```

```python
fig, (ax1) = plt.subplots(1,1, figsize = (20,10))
ax1.plot(data.index, data['TIPO CAMBIO COMPRA'], '-',  color = 'teal', label = 'TCC')
ax1.plot(data.index, data['TIPO DE CAMBIO VENTA'], '-',  color = 'aquamarine', label = 'TCV')
ax1.axhline(data['TIPO CAMBIO COMPRA'].mean(), color='goldenrod', linestyle='--', label='E(TCC)')
ax1.axhline(data['TIPO DE CAMBIO VENTA'].mean(), color='khaki', linestyle='--', label='E(TCV)')
ax1.set(xlabel= 'Periodicity', ylabel= "Price")
ax1.set_title('Evolution of Exchange Rate Costa Rica, Series 01-01-2021 / 01-12-2023')
ax1.legend(loc='best', fontsize = 10)
plt.subplots_adjust(top=0.84)
plt.show()
```

```python
DeltaCompra = data['TIPO CAMBIO COMPRA'][0]-data['TIPO CAMBIO COMPRA'][-1]
DeltaVenta = data['TIPO DE CAMBIO VENTA'][0]-data['TIPO DE CAMBIO VENTA'][-1]
print(DeltaCompra,".........",DeltaVenta)
```
